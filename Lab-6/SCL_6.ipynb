```
import matplotlib.pyplot as plt 
# creating the dataset 
data = {'EC':130, 'CS':70, 'ME':190, 'CIVIL':130, 'B.Arch':40} 
x = list(data.keys())  			#acquiring elements of dict
y = list(data.values()) 

# creating the bar plot 
plt.bar(x, y,color=['green','orange','yellow','red','blue'],width=0.3) 
plt.xlabel("Departments") 
plt.ylabel("No. of students") 
plt.title("Students admitted in different departments") 
plt.show() 
```
