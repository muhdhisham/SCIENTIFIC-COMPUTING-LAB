```
import numpy as np                  #import numpy
import matplotlib.pyplot as plt     #import matplotlib.pyplot

T=20                                
x=[]                                #defining a empty list
y=[]
i=5
time=np.arange(0,100,0.01)
def f(n):                                                                           #define function that returs the fourier series
  return ((-1)**(n+1))*(4/(np.pi*((2*n)-1)))*np.cos((2*np.pi*((2*n)-1)*t)/T)
for t in time:                                                     
  for n in range(1,i+1):                                                             #for loop for getting functions f(1),f(2)..at a particular time
    y.append(f(n))                                                                  #getting the functions to list
  x.append(np.sum(y))                                                               #obtaining the sum given functions
  y=[]                                                                              #emptying y
 
 
plt.plot(time,x)                     #plotting
plt.show()
```
