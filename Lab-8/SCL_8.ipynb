Q1

```
import numpy as np                    # importing numpy
import pandas                         # importing pandas
from pandas import DataFrame          # importing data frame from pandas
C = {'time': [],                      # creating dictionary
'sin t':[],
'cos t': [],
}
for i in range(50):                   # appending datas to the respective columns
 C['time'].append(i)
 C['sin t'].append(np.sin(i))
 C['cos t'].append(np.cos(i))
df = DataFrame(C, columns= ['time','sin t','cos t'])              # creating a data frame
export_csv = df.to_csv ('scl.csv',index = None, header=True)

import csv
with open('scl.csv','r') as csvfile:
  reader=csv.reader(csvfile,skipinitialspace=True)
  for row in reader:
    print(row)
```
