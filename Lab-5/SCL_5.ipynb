```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dx/dt
def f(x,t):
    dxdt = -2*x 
    return dxdt

# initial condition
x0 = 1

# time points
t = np.linspace(0,5)

# solve ODE
y = odeint(f,x0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
                                      



```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def f(U, t):
    # Here U is a vector such that y=U[0] and z=U[1]. This function should return [y', z']
    return [U[1], -1000*U[1] -5000*np.exp(-1*t)- 1000000000*U[0]]
U0 = [0, 5000]                   #initial conditions
t = np.arange(0, 0.01,0.00005)   #time points
Us = odeint(f, U0, t)            #solving
xs = Us[:,0]
plt.xlabel("x")
plt.ylabel("y")
plt.plot(t,xs);                  #plotting
```
